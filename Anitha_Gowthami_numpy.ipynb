### Q4. Create a 2-D numpy array of random integers between 1 to 100 with shape 5*4. Now perform following operations: 
• Print only the even numbers from this array 
• Print only the odd numbers from this array 
• Print row wise sum of all elements 
• Print column wise sum of all elements 
• Convert this 2-D array into 3-D array of shape 
2*2*5

```python
import numpy as np
Q4_array = np.random.randint(1,100,size =(5,4))

 #1. Print only the even numbers from this array 
print("-------------Even Numbers--------------")
for i in range(Q4_array.shape[0]):
    for j in range(Q4_array.shape[1]):
        if Q4_array[i,j] % 2 == 0:            
            print((i,j),":",Q4_array[i,j])

#2. Print only the odd numbers from this array 
print("-------------Odd Numbers--------------")
for i in range(Q4_array.shape[0]):
    for j in range(Q4_array.shape[1]):
        if Q4_array[i,j] % 2 != 0:            
            print((i,j),":",Q4_array[i,j])

#3. Print row wise sum of all elements 
row_sum = np.sum(Q4_array, axis = 1)

#4. Print column wise sum of all elements
col_sum = np.sum(Q4_array, axis = 0)

print(f"\nrow wise sum:\n{row_sum}\n\n column wise sum:\n{col_sum}")    

#5. Convert this 2-D array into 3-D array of shape 225
print(f" \n\n  2-D to 3-D array \n ===================\n {Q4_array.reshape(2,2,5)}")
```

### Q5. Create a 2-D numpy array of random integers between 1 to 100 with shape 5*4. Now perform following operations: 
• Swap first row with third row 
• Swap second column with fourth column. 
• Replace the values less than 50 with zero. 
• Convert this 2D array into 1-D array 

```python
import numpy as np
Q5_array = np.random.randint(1,100,size = (5,4))

# Swap first row with third row 
swapped = Q5_array.copy()
swapped[[0,2]] = swapped[[2,0]]

print("\n\nSwap first row with third row\n ")
print(f"{'Before Swap':<25}{'After Swap':<30}")
for a, b in zip(Q5_array, swapped):
    print(f"{a}\t\t {b}")

print("Original array size :",Q5_array.shape)
print("swapped array size :",swapped.shape)

#=================================================
#2 Swap second row with fourth row 
swapped = Q5_array.copy()
swapped[[1,3]] = swapped[[3,1]]

print("\n\nSwap second row with fourth row\n")
print(f"{'Before Swap':<25}{'After Swap':<30}")
for a, b in zip(Q5_array, swapped):
    print(f"{a}\t\t {b}")

print("Original array size :",Q5_array.shape)
print("swapped array size :",swapped.shape)

#=================================================
# 3 Replace the values less than 50 with zero
print("\n\nReplace the values less than 50 with zero\n")
replced_Q5 = Q5_array.copy()
for i in range(replced_Q5.shape[0]):
    for j in range(replced_Q5.shape[1]):
        if replced_Q5[i,j] < 50 :
            replced_Q5[i,j] = 0
        else:
            replced_Q5[i,j] = replced_Q5[i,j]
print(replced_Q5)

#===========================================================
#Convert this 2D array into 1-D array
Q5_1D = Q5_array.flatten()
print("\n\nConvert this 2D array into 1-D array\n")
print("Q5_1D:", Q5_1D)
```

### Q6. Explain the concept of stacking in numpy with examples. 
How is it different from concatenation. Give any 3 
differences between them. 


1. Stacking means joining two or more arrays along a new axis — that is, it adds a new dimension
   while combining arrays.
   np.vstack() → vertical stack (axis=0, like rows)
   np.hstack() → horizontal stack (axis=1, like columns)
   np.dstack() → depth stack (axis=2, 3D stacking)
2. Concatenation means joining arrays along an existing axis — it does not create a new axis.

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Stack along a new axis (axis=0)
stack1 = np.stack((a, b), axis=0)
print(stack1)
print(stack1.shape)

stack2 = np.stack((a, b), axis=1)
print(stack2)
print(stack2.shape)
```

```python
a = np.array([[1, 2],
              [3, 4]])
b = np.array([[5, 6]])

# Concatenate along existing axis 0
concat1 = np.concatenate((a, b), axis=0)
print(concat1)
print(concat1.shape)
```

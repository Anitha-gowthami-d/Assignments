### 4. Write code to create a user defined function named Calculator.
●	User will pass 3 parameters in Calculator function – int1, int2, task
●	The task string will be ‘add’, ‘sub’, ‘mul’, ‘div’. Perform Addition, Subtraction, Multiplication and Division on the two numbers(int1, int2) respectively. Make sure the task parameter is case insensitive (meaning that even if user passes ‘aDd’ as command it should perform addition on two numbers).
●	Define logics for addition, subtraction, multiplication and division in Calculator function that take in two numbers and return another number after performing their respective operations.
Eg: Calculator(2,5,ADD) should return 7
       Calculator(7,1,SuB) should return 6


```python
int1 = input("Enter your first value in numbers: ")
int2 = input("Enter your second value in numbers: ")
task = input("Enter your operator(‘add’, ‘sub’, ‘mul’, ‘div’) :")

def user_calculator(int1,int2,task):
    try:
        a = int(int1)
        b = int(int2)
    except Exception:
        print("Please Enter valid numbers")
        return
        
    operation_task = str(task).lower()
    try:
        if operation_task in ['addition','add','+']:
            result = a+b
        elif operation_task in ['subtraction', 'sub','-']:
            result = a-b
        elif operation_task in ['multiplication', 'mul','prod','product','x','*']:
            result = a*b
        elif operation_task in ['division', 'div','/']:
            result = a/b
        else:
            print("Enter valid input operator")
    except ZeroDivisionError:
        print("Division by zero is not possible")
        return
        
    return result
user_calculator(int1,int2,task)
```

### 5. Write a Function to check if the year number is a leap year or not. Print all Leap Years of 21st Century.

```python
def leap_year(start, stop):
    for i in range(start, stop + 1):
        if (i % 4 == 0 and (i % 100 != 0 or i % 400 == 0)):
            print(i, "is a leap year")
leap_year(2000, 2100)
```

### Q8:Write a Function to take a list from user and return 2 lists:
-	1st list should contain all even indexed numbers.
-	2nd list should contain all odd indexed numbers


```python
list1 = input("Enter your input list seperated by comma:")

def indexed_list(input_list):
    a = str(input_list).split(",")
    list_even = []
    list_odd = []
    for i in range(0,len(a)):
        if i % 2 == 0:
            list_even.append(a[i])
        else:
            list_odd.append(a[i])
    return list_even, list_odd

two_lists = indexed_list(list1)
print("Even_Index:",two_lists[0],"Odd_Index:",two_lists[1],sep ="\n")
```

### Q9. Create a function longest_word(sentence) that takes a sentence as input and returns  the  longest word in  the  sentence. If two or  more words have the  same length, return the first one.

```python
sentence =input("Enter your sentence here:")
ss = sentence.split()
ss_count = ""
for i in ss:
    if len(i) <= len(ss_count):
        ss_count = ss_count
    else:
        len(i) > len(ss_count)
        ss_count = i
        
print(ss_count)
        
```
